## **Big O and time complexity**



**Analogy**: You've got a file on a hard drive and you need to send it to your friend who lives across the country. You need to get the file to your friend as fast as possible. How should you send it?

If it's a small file, you might use email, FTP, or some other means of electronic transfer.

But what if the file were really, really large? It might be faster to physically deliver it via plane. If your file is that urgent (and cost isn't an issue), you might just want to do that. It would take 5 -10 hours to get to an airport, hop on a flight, and then deliver the file to your friend. But a one-terabyte (1 TB) file could take more than a day to transfer electronically.

What if there were no flights, and instead you had to drive across the country? Even then, for a really huge file, it would be faster to drive. 

This is what the **concept of asymptotic runtime**, or **big O** time, means. 

- Electronic Transfer: O(s), where s is the size of the file. This means that the time to transfer the file increases linearly with the size of the file.
- Airplane Transfer: O(1) with respect to the size of the file. As the size of the file increases, it won't take any longer to get the file to your friend. The time is constant. 

There are many more runtimes than this. Some of the most common ones are $O(log N)$, $O(N)$, $O(N log N)$, $O(N^2)$ and $O(2^N)$. There's no fixed list of possible runtimes, though. 

**In academia**, big O describes an **upper bound** on the time. An algorithm that prints all the values in an array could be described as $O(N)$, but it could also be described as $O(N^2)$, $O(N^3)$, or $O(2N)$ (or many other big O times).

In academia, $\Omega$ is the equivalent concept but for **lower bound**. Printing the values in an array is $O(N)$ as well as $O(log N)$ and $O(1)$. After all, you know that it won't be faster than those runtimes. 

In academia, $\Theta$ means both $O$ and $\Omega$. That is, an algorithm is $0(N)$ if it is both $O(N)$ and $\Omega( N)$. $\Theta$ gives a tight bound on runtime.

In industry (and therefore in interviews), people seem to have merged $\Theta$ and $O$ together. Industry's meaning of big O is closer to what academics mean by $\Theta$, in that it would be seen as incorrect to describe printing an array as $O(N^2)$. Industry would just say this is $O(N)$. 

**Best Case, Worst Case, and Expected Case**

We can actually describe our runtime for an algorithm in three different ways. Best, worst, and expected cases describe the big O (or big theta) time for particular inputs or scenarios. 

We rarely ever discuss best case time complexity, because it's not a very useful concept. After all, we could take essentially any algorithm, special case some input, and then get an $O(1)$ time in the best case. For many-probably most-algorithms, the worst case and the expected case are the same.



**Space complexity**

Space complexity is a parallel concept to time complexity. If we need to create an array of size n, this will require $O(n)$ space. If we need a two-dimensional array of size nxn, this will require $O(n^2)$ space. Stack space in recursive calls counts, too. Therefore, recursive algorithms are always memory-heavy.



**Constants**

Big O just describes the rate of increase. For this reason, we drop the constants in runtime. An algorithm that one might have described as $O(2N)$ is actually $O(N)$.

**Amortized runtime**

An `Array List`, or a dynamically resizing array, allows you to have the benefits of an array while offering flexibility in size. Lists in Python are implemented that way.

When the array hits capacity, the `Arraylist` class will create a new array with double the capacity and copy all the elements over to the new array.

How do you describe the runtime of insertion? This is a tricky question. The array could be full. If the array contains N elements, then inserting a new element will take O(N) time. You will have to create a new array of size 2N and then copy N elements over.

However, we also know that this doesn't happen very often. As we insert elements, we double the capacity when the size of the array is a power of 2. So after X elements, we double the capacity at array sizes 1, 2, 4, 8, 16, ... , X.

That doubling takes, respectively, 1, 2, 4, 8, 16, 32, 64, ... , X copies. This is roughly 2X. Therefore, X insertions take O(2X) time. The amortized time for each insertion is O(1).

**Contains dublicate (Easy) - Microsoft**

https://leetcode.com/problems/contains-duplicate/

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

 
**Example 1:**

- Input: nums = [1,2,3,1]
- Output: true

**Example 2:**

- Input: nums = [1,2,3,4]
- Output: false

**Example 3:**

- Input: nums = [1,1,1,3,3,4,3,2,4,2]
- Output: true
 

Constraints:

- 1 <= nums.length <= 105
- -109 <= nums[i] <= 109

```python
# solution 1 See https://www.youtube.com/watch?v=3OamzN90kPg

'''We define a set (doesn't contains dublicates) and loop over all elements in
nums. If the current element is already in the set, we found a dublicate.'''

from typing import List

def containsDuplicate(self, nums: List[int]) -> bool:
        
        hash = set()
        
        for n in nums:
            if n in hash:
                return True
            hash.add(n)
        return False

# solution 2 as one-liner

'''The first solution is not really taking advantage of set's main property (no
dublicates allowed). We can make a set out of nums and compare the length of nums
with the list cleaned of dublicates.'''

from typing import List

def containsDuplicate(self, nums: List[int]) -> bool:
  return len(set(nums)) != len(nums)
```

**Valid Anagram (Easy) - Uber**

https://leetcode.com/problems/valid-anagram/

Given two strings s and t, return true if t is an anagram of s, and false otherwise.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

**Example 1:**

- Input: s = "anagram", t = "nagaram"
- Output: true

**Example 2:**

- Input: s = "rat", t = "car"
- Output: false
 

**Constraints:**

- 1 <= s.length, t.length <= 5 * 104
- s and t consist of lowercase English letters.
 

**Follow up:** What if the inputs contain Unicode characters? How would you adapt your solution to such a case?

```python
# solution 1 (spontaneous solution)

'''We can sort s and t, and compare the sorted strings. But we have to keep sorting
complexity in mind which is about O(n log n).'''

def isAnagram(self, s: str, t: str) -> bool:
        
    return sorted(s) == sorted(t)

# solution 2 See https://www.youtube.com/watch?v=9UtInBqnCgA

'''The goal is to count occuriences of strings contained in s and t. We use two
dicts (hash tables) for this. We iterate through s an t for that. Finally, we
check whether the hash tables are equal. Time complexity is O(s+t). We create
two hash tables.'''

def isAnagram(self, s: str, t: str) -> bool:
    
    s_freq, t_freq = {}, {}
    
    for key in s:
        # if the corresponding key is not in freq dict, use 0 as default value
            s_freq[key] = s_freq.get(key, 0) + 1
            
    for key in t:
            t_freq[key] = t_freq.get(key, 0) + 1
            
    if s_freq == t_freq:
        return True
    else:
        return False
```

**Two sum (Easy) - Facebook**

https://leetcode.com/problems/two-sum/

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

**Example 1:**

- Input: nums = [2,7,11,15], target = 9
- Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

**Example 2:**

- Input: nums = [3,2,4], target = 6
- Output: [1,2]

**Example 3:**

- Input: nums = [3,3], target = 6
- Output: [0,1]
 

**Constraints:**

- 2 <= nums.length <= 104
- -109 <= nums[i] <= 109
- -109 <= target <= 109
- Only one valid answer exists.
 

**Follow-up:** Can you come up with an algorithm that is less than O(n2) time complexity?

```python
# solution 1 (spontaneous solution)

'''Simplest solution is a nested for-loop where we are looping twice over the
indexes and check the sum of the corresponding list elements. We have to fulfill
two conditions: the sum of the two elements have to equal to the target AND the
both indexes have to be different. Very slow solution with O(n**2) complexity'''

from typing import List

def twoSum(self, nums: List[int], target: int) -> List[int]:
    for i in range(len(nums)):
        for j in range(len(nums)):
            if i != j and (nums[i] + nums[j] == target):
                return [i, j]

# solution 2 See https://www.youtube.com/watch?v=KLlXCFG5TnA

'''We create a dict and loop over the nums list and are addint elements as keys
and the corresponding indexes as values. Main trick: on each iteration we
additionaly compute the difference between the current element and the target.
Then we check in the dict whether we already encountered an element which would
add up to the target with the current element. If we have, we can instantly
retrieve that element's index from the dict. Saving and retrieving from a hash
has constant time complexity, so we are left with O(n) for iterating once over
the list.'''

from typing import List

def twoSum(self, nums: List[int], target: int) -> List[int]:
        
    hash_dict = {} # value: index
    
    for i, n in enumerate(nums):
        difference = target - n
        if difference in hash_dict:
            return [hash_dict[difference], i]
        hash_dict[n] = i # key: number
```

**Groups of Anagrams (Medium) - Amazon**

https://leetcode.com/problems/group-anagrams/

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

**Example 1:**

- Input: strs = ["eat","tea","tan","ate","nat","bat"]
- Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

**Example 2:**

- Input: strs = [""]
- Output: [[""]]

**Example 3:**

- Input: strs = ["a"]
- Output: [["a"]]
 

**Constraints:**

- 1 <= strs.length <= 104
- 0 <= strs[i].length <= 100
- strs[i] consists of lowercase English letters.

```python
# solution 1 see https://www.youtube.com/watch?v=vzdNOK2oB2E

'''Refer to the basic anagram problem. Using sorting results in O(m*n log n)
apparently, but I find this solution still more compeling..? We loop through the
strs list, sort each string, and compile a hash table. Resulting dict:
{'a', 'e', 't':['eat', 'tea']}'''

from typing import List

def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        
    results = {}

    for str in strs:
        if tuple(sorted(str)) in results:
            results[tuple(sorted(str))].append(str)
        else:
            results[tuple(sorted(str))] = [str] # only tuples work as keys

    return list(results.values())

# solution 2

'''This solution doesn't use sorting, but character counts. This only woks if we
assume that we have only 26 latin characters. Also, it uses a rather convoluted
implementation of one-hot encoding for the count vector where each character has
to be converted to the unicode value and normalized with respect to 'a'.'''

from typing import List

def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        
    results = {}

    for str in strs:
        
        # a one-hot vector of chars count
        count = [0]*26 # a...z
        
        for char in str:
            # use normalized unicode number as index
            count[ord(char)-ord('a')] += 1

        if tuple(count) in results:
            results[tuple(count)].append(str)
        else:
            results[tuple(count)] = [str]

    return list(results.values())
```

**Top k frequent elements (Medium) - Amazon**

https://leetcode.com/problems/top-k-frequent-elements/

Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

**Example 1:**

- Input: nums = [1,1,1,2,2,3], k = 2
- Output: [1,2]

**Example 2:**

- Input: nums = [1], k = 1
- Output: [1]
 

**Constraints:**

- 1 <= nums.length <= 105
- k is in the range [1, the number of unique elements in the array].

It is guaranteed that the answer is unique.

**Follow up:** Your algorithm's time complexity must be better than O(n log n), where n is the array's size.

```python
# solution 1 (came up with the idea, had to look up the syntax)

'''We first create a frequency dictionary. Sort it by value and return top k
elements from the sorted list. For relevant Python syntax see
https://stackoverflow.com/questions/12266617/finding-top-k-largest-keys-in-a-dictionary-python
The solution has a O(n log n) complexity.'''

def topKFrequent(self, nums: List[int], k: int) -> List[int]:
  
  freq_dict = {}

  for num in nums:
    freq_dict[num] = freq_dict.get(num, 0) + 1
  
  return sorted(freq_dict, key=freq_dict.get, reverse=True)[:k]

# solution 2 with O(n log k) complexity 

'''The idea remains the same, but now we're using a heap so we don't have to sort
the entire dict, but only the top k elements. See
https://www.geeksforgeeks.org/heap-queue-or-heapq-in-python/ about heapq module
and https://stackoverflow.com/questions/2243542/how-to-efficiently-get-the-k-bigger-elements-of-a-list-in-python
for nlargest specificaly'''

import heapq

def topKFrequent(self, nums: List[int], k: int) -> List[int]:
  
  freq_dict = {}

  for num in nums:
    freq_dict[num] = freq_dict.get(num, 0) + 1
  
  return heapq.nlargest(k, freq_dict, key=freq_dict.get)

# solution 3 with O(n) using bucket sort see https://www.youtube.com/watch?v=YPTqKIgVk-k
# !! this was the slowest solution in leetcode !! and has objectively the worst
# memory usage

def topKFrequent(self, nums: List[int], k: int) -> List[int]:
  
  freq_dict = {}
  
  '''The index of the bucket list will correspond to the number of occurences of
  each digit inside the nums list. Bucket list has to be a list of lists because
  some digits might have the same number of occurences.'''
  bucket_list = [[]]  

  for num in nums:
    freq_dict[num] = freq_dict.get(num, 0) + 1
  
  for num, count in freq_dict.items():
    bucket_list[count].append(num)
  
  result = []

  for i, num_list in enumerate(bucket_list[:-1]):
    for num in bucket_list[i]:
      result.append(num)

      if len(result) == k:
        return result
```

```python
def productExeptSelf(nums: List[int]) -> List[int]:
  # initialize the output array
  output_array = 1*len(nums)

  # set prefix counter to 1
  prefix = 1
  # forward pass filling up the array with prefixes
  # 0 1 2 3 (index list)
  for i, element in enumerate(nums):
    output_array[i] = prefix
    # update prefix
    prefix *= nums[i]

  # set postfix counter to 1
  postfix = 1
  
  # applying enumerate to a reversed list doesn't make sense
  # !! enumerate objects are not reversible !!
  # for i, element in enumerate(reversed(nums)): # slower enumerate(nums[::-1])
  
  # some fuckery here with the range function to get
  # 3 2 1 0 (index list)
  for i in range(len(nums)-1, -1, -1):
    # multiplying prefixes with postfixes
    output_array[i] *= postfix
    # update postfix
    postfix *= nums[i]
  
  return output_array

```

```python
'''For example, cell [3,3] lies in the [1,1] sub-grid. We can convert the original
index by dividing the it by the size of a sub-grid (3) using integer division.

This results in dividing and flooring the result. For example, 3//3 = 1 or
4//3 = 1 or 5//3 = 1. That means that each key of the third hash table will be
[row//3, column//3].

'''
import collections

def isValidSudoku(self, board: List[List[str]]) -> bool:
  columns = collections.defaultdict(set)
  rows = collections.defaultdict(set)
  squares = collections.defaultdict(set) # key = (r//3, c//3)

  for row in range(9):
    for col in range (9):
      if board[row][col] == ".": # empty spot
        continue
      
      # check for dublicates in the row hash table
      if (board[row][col] in rows[row]):
        return False

      # check for dublicates in the column hash table
      if (board[row][col] in columns[col]):
        return False
      
      # check for dublicates in the subgrid hash table
      if (board[row][col] in squares[(row//3,col//3)]):
        return False

      # in a single step
      # if (board[row][col] in rows[row] or
      #     board[row][col] in columns[col] or
      #     board[row][col] in squares[(row//3,col//3)]):
      #   return False

      # update all hash tables with current value
      columns[col].add(board[row][col])
      rows[row].add(board[row][col])
      squares[(row//3,col//3)].add(board[row][col])

  return True

```

```python
'''That is we have first to identify how many sequences there are in the input
array without sorting it. To find the starting point of a potential sequence, we
have to check whether each number has a "neighbor" on its left side.

First, we convert the array to a set(?). Then we loop through the set and check for
each number x whether the set also contains x-1. If it doesn't, we are at the
start of sequence. Now, we check whether we can continue with the sequence by
checking x+1. Once we cannot find x+1, we move on to the next number in the set.  

Finally, we compare the length of the created sequences.

The final solution has time and memory complexity of O(n).
'''
```

```python
def longestConsecutive(nums: List[int]) -> int:
  num_set = set(nums)

  max_seq_length = 0

  # iterate through the array
  for num in nums:
    # check it the number is a start of a sequence
    if (num-1) not in num_set:
     current_seq_length = 0
     while (num+current_seq_length) in num_set:
       current_seq_length += 1
     max_seq_length = max(current_seq_length, max_seq_length)
  
  return max_seq_length


```

**Two Sum II (Medium) - Amazon**

https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/

Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.


Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2.

The tests are generated such that there is **exactly one solution**. You may not use the same element twice.

Your solution must use only constant extra space.

 

**Example 1:**

- Input: numbers = [2,7,11,15], target = 9
- Output: [1,2]

**Explanation:** The sum of 2 and 7 is 9. Therefore, index1 = 1, index2 = 2. We return [1, 2].

**Example 2:**

- Input: numbers = [2,3,4], target = 6
- Output: [1,3]

**Explanation:** The sum of 2 and 4 is 6. Therefore index1 = 1, index2 = 3. We return [1, 3].

**Example 3:**

- Input: numbers = [-1,0], target = -1
- Output: [1,2]

**Explanation**: The sum of -1 and 0 is -1. Therefore index1 = 1, index2 = 2. We return [1, 2].
 

**Constraints:**

- 2 <= numbers.length <= 3 * 104
- -1000 <= numbers[i] <= 1000
- numbers is sorted in non-decreasing order.
- -1000 <= target <= 1000

The tests are generated such that there is **exactly one solution**.

```python
# see https://www.youtube.com/watch?v=cQ1Oz4ckceM

# solution 1: brute force
'''Simplest solution is a nested for-loop where we are looping twice over the
indexes and check the sum of the corresponding list elements. We have to fulfill
two conditions: the sum of the two elements have to equal to the target AND the
both indexes have to be different. Very slow solution with O(n**2) complexity
Simple addition: for each consecutive loop ignore elements with higher sum than
the target. '''

# solution 2 with a hash table (extra memory)
def twoSum(self, numbers: List[int], target: int) -> List[int]:
    hash_dict = {} # value: index

    for i, n in enumerate(numbers):
        difference = target - n
        if difference in hash_dict:
            return [hash_dict[difference], i+1]
        hash_dict[n] = i+1 # key: number

# solution 3 with two pointers (works only is the array is sorted)
''' We define left and right pointers. At each iteration, we check the sum of
both elements at the pointers. If the sum is bigger than target, we decrease the
right pointer (the array is sorted!). On the other hand, if the sum is too small,
we increase the left pointer. We are garanteed to find a solution. The time
complexity remains O(n), but we don't use any extra memory like in the solution
with a hash table.'''

def twoSum(self, numbers: List[int], target: int) -> List[int]:
  l, r = 0, len(numbers)-1

  while l < r:
      curSum = numbers[l] + numbers[r]

      if curSum < target:
          l += 1
      elif curSum > target:
          r -= 1
      else:
          return [l+1, r+1] 
```

**Constraints:**

- -105 <= nums[i] <= 105

```python
# see https://www.youtube.com/watch?v=jzZsG8n2R9A

# solution 1: brute force
'''We set the first unique element as A and the last element of the array as
C. Now, we look at all elements between A and C as a potential B. Once we have
checked all possible combinations, we increase the pointer at A and
look for another combination. This solution doesn't account for dublicates and
has O(n**3) complexity in the worst case.'''

# solution 2: with pointers
'''To filter out dublicate triplets, we sort the array first. By doing this, we
can easily skip numbers in the A position which would potentially lead to the
same solution. Then, we set the first unique element in the sorted array as A
and solve for B and C with the same solution we have used for Two Sum II.
Different from Two Sum II, we have to update the pointers.

The time complexity ia O(n log n) for sorting and O(n**2) because we have an outer
loop for setting A and an inner loop for finding matching B and C. That means that
the overall complexity is O(n**2). The space complexiy depends on the implementation
of the sorting part.'''

def threeSum(self, nums: List[int]) -> List[List[int]]:
  nums.sort()
  solutions = []
  
  for i, n in enumerate(nums):
      
      if i > 0 and n == nums[i-1]:
          continue
      
      l, r = i+1, len(nums)-1

      while l < r:
          curSum = n + nums[l] + nums[r]

          if curSum < 0:
              l += 1
          elif curSum > 0:
              r -= 1
          else:
              solutions.append([n, nums[l], nums[r]])
              # why do we have do this?
              l += 1
              while nums[l] == nums[l-1] and l < r:
                  l+=1
  
  return solutions
```

```python
# see https://www.youtube.com/watch?v=1pkOgXD63yU

# solution 1 with two pointers
'''The general idea is to buy low and sell high. We can set the left pointer for
day X, and right pointer for day X+1. We track the difference between the amount
at right pointer and the amount at left pointer. If the value at the right pointer
is lower than at the left pointer, we set the left pointer to the position of the
right pointer. As long as the value at the right pointer stays below the value at
the left pointer, we don't change the position of the left pointer, but only
increase the right pointer. The time complexity is O(n) and memory complexity is
O(1) as we don't need any extra memory.'''

def maxProfit(self, prices: List[int]) -> int:
  l, r = 0, 1
  max_profit = 0

  while r < len(prices):
      if prices[l] < prices[r]:
          profit = prices[r] - prices[l]
          max_profit = max(max_profit, profit)
      else:
          l = r
      r += 1
  
  return max_profit
```

```python
# see https://youtu.be/XIdigk956u0

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: ListNode, list2: ListNode) -> ListNode:
        dummy = ListNode()
        tail = dummy

        while list1 and list2:
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        if list1:
            tail.next = list1
        elif list2:
            tail.next = list2

        return dummy.next
```

- Output: 2

1. 1 step + 1 step
2. 2 steps

**Example 2:**

- Input: n = 3
- Output: 3

**Explanation:** There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
 

**Constraints:**

- 1 <= n <= 45

image.png

```python
# see https://www.youtube.com/watch?v=Y0lT9Fck7qI

# brute force
- at each step, we have to decide to take either a single or two steps
- drawing the desicion as a branching tree we arrive at the number of possible
solutions
- use could apply recurion and a depth-first search
- however, we see redundant computations (4 always leads to 5 in a single step)
- the sub tree for 2 is also identical
- time complexity is 2 (binary choice at each step)**5 (goal)
- O(2**n)
```
