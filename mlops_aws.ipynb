## EC2 instances (virtual machines) provide GPUs

Create an EC2 (https://sagemaker-examples.readthedocs.io/en/latest/) instance:
- search for EC2 and go to the EC2 Dashboard
- increase the limit of the running/requested instances (G/P instances 16vCPUs, standard 64 vCPUs) - takes about 2 days to get approved
- launch instance: instances/Lauch instance on the top-right corner

- choose Amazon Machine Image (OS): AWS Deep Learning AMI (Ubuntu) - includes Anaconda, ML packages (TensorFlow, PyTorch), CUDA, CuDNN, drivers
- choose instance type: for details see https://aws.amazon.com/ec2/instance-types/
- t2.micro (no GPU, one core 1 GB RAM) - free; suitable for testing
- c5.xlarge (no GPU, 4 cores 8 GB RAM) - basic code development
- g4dn.xlarge (16 GB T4 GPU, 4 cores 16 GB RAM) - code development and model training + 125GB of storage which can be used for training data
- p3.2xlarge (16 GB V100 Tesla GPU, 8 core 61 GB RAM) - accelerated training of large models - expensive!
- click next to configure instance details: on-demand instance vs persistent spot instance vs requested spot instance (spot instances cheaper, stop behavior is different), you can also specify price limits to stop an instance
- select an existing security group: ssh access
- create/use a security key pair
- to connect to the EC2 instance use your terminal: ssh -i /path_to_key-file ubuntu@my-instance-public-ip-address
- get the IP from the EC2 Dashboard
- MobaXTerm on Windows
- source activate pytorch_latest_p37 to activate vitual environments
- htop to see RAM and CPU usage
- nvidia-smi
- stop/terminate an instance on the spot request dashboard or instance dashboard

- create lanuch templates to speed up the launch process
- select a running instance, go to actions in the top-right corner, choose Image and templates, create template
- select desired options and save it under a name
- lauch instance from template
- install own anaconda (see https://chen-zhe.github.io/blog/p/aws-user-notes-for-deep-learning/)
- wget link-to-file
- chmod +x install_script.sh 
- run the install script
- source .bashrc to start the base anaconda environment
- conda install jupyterlab
- for set up see https://www.youtube.com/watch?v=25VuoasPues minute 7
- can then access jupyter lab via browser; possible to use the terminal there which launches on the EC2 instance automatically

SageMaker enables developers to operate at a number of levels of abstraction when training and deploying machine learning models. 

At its highest level of abstraction, SageMaker provides pre-trained ML models that can be deployed as-is. In addition, SageMaker provides a number of built-in ML algorithms that developers can train on their own data.

Further, SageMaker provides managed instances of TensorFlow and Apache MXNet, where developers can create their own ML algorithms from scratch.

Regardless of which level of abstraction is used, a developer can connect their SageMaker-enabled ML models to other AWS services, such as the **Amazon DynamoDB database** for structured data storage, AWS Batch for offline batch processing, or **Amazon Kinesis for real-time processing**.
